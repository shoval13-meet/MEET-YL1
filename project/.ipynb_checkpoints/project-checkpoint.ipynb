```python
import csv
from pylab import *

p = open("environmental.csv" , "rb")
s = csv.reader(p)

index = 0
ozone = []
wind = []

for row in  s:
    if index != 0:
        r = row
        m = map(float, r[1:])
        ozone.append(m[0])
        wind.append(m[1])
        
    index += 1

p.close()
print ozone,wind  

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(ozone,wind)
ax.set_title("relationship between the ozone and the wind")
ax.set_xlabel("ozone")
ax.set_ylabel("wind")

savefig("mygraph.png")
```

conclusion : although i expected to see a connection between the wind and the ozone ,there is no connection between them.
